```r
library(caTools)

split = sample.split(sg_final$predictor, SplitRatio = .8)
training_data = subset(sg_final, split == TRUE)
test_data = subset(sg_final, split == FALSE)
nrow(training_data)
nrow(test_data)
ncol(test_data)
```

```r
library(e1071)
svm_model<-svm(predictor~.,data=training_data)
summary(svm_model)
```

```r
pred <- predict(svm_model,test_data[,-328])
cm=table(pred,test_data$predictor)

accuracy <- sum(diag(cm))/sum(cm)
```
